# MatPlotLib

```python
# Importing required libraries
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

# 0 : Chart Types

```python
# Load a dataset from seaborn package
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
map_data = px.data.gapminder().query("year==2007")
```

## Line Plot
### Pros:
1. Shows the trend or change in data over time.
2. Suitable for displaying continuous data.

### Cons:
1. Overlapping lines can make it difficult to interpret the plot.
2. Not ideal for displaying categorical data.

```python
plt.figure(figsize=(8, 4))
sns.lineplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Line Plot')
plt.show()
```

## Bar Plot
### Pros:
1. Great for comparing categorical data.
2. Easy to understand and interpret.

### Cons:
1. Can become cluttered with too many categories.
2. Usually limited to representing one variable on the y-axis.

```python
plt.figure(figsize=(8, 4))
sns.barplot(data=tips, x='day', y='total_bill', hue='sex')
plt.title('Bar Plot')
plt.show()
```

## Pie Chart
### Pros:
1. Suitable for displaying parts of a whole.
2. Easy to interpret percentages.

### Cons:
1. Not ideal for comparing multiple categories.
2. Can be misleading when the number of categories is large.

```python
plt.figure(figsize=(6, 6))
plt.pie(iris['species'].value_counts(), labels=iris['species'].unique(), autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

## Histogram
### Pros:
1. Visualizes the distribution of continuous or discrete data.
2. Provides insights into the data's central tendency and spread.

### Cons:
1. Choosing an inappropriate number of bins can distort the interpretation.
2. Not suitable for displaying categorical data.

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=tips, x='total_bill', bins=20, kde=True)
plt.title('Histogram')
plt.show()
```

## Pair Plot
### Pros:
1. Displays the pairwise relationships between multiple variables.
2. Useful for identifying correlations and patterns in the data.

### Cons:
1. Can become crowded and complex with a large number of variables.

```python
sns.pairplot(data=iris, hue='species')
plt.title('Pair Plot')
plt.show()
```

## Scatter Plot
### Pros:
1. Visualizes the relationship between two continuous variables.
2. Useful for identifying trends, clusters, or outliers.

### Cons:
1. Limited to displaying two variables at a time.
2. Difficult to interpret when there are many datapoints that overlap.

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Scatter Plot')
plt.show()
```

## Dot Plot
### Pros:
1. Shows the distribution of a quantitative variable across categories.
2. Efficiently displays data points and their positions.

### Cons:
1. Can become cluttered when there are too many categories or data points.
2. Limited to displaying one variable on the y-axis.

```python
plt.figure(figsize=(8, 4))
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.title('Dot Plot')
plt.show()
```

## Count Plot
### Pros:
1. Shows the count of observations in each category.
2. Provides a clear overview of the distribution of categorical data.

### Cons:
1. Limited to displaying one variable.
2. Does not represent the actual values of the observations.

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=tips, x='day', hue='sex')
plt.title('Count Plot')
plt.show()
```

## Box Plot
### Pros:
1. Displays the distribution of a continuous variable across categories.
2. Reveals information about median, quartiles, and outliers.

### Cons:
1. May not be suitable for small datasets.
2. Does not show the actual data points.

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=iris, x='species', y='sepal_width')
plt.title('Box Plot')
plt.show()
```

## Heatmap
### Pros:
1. Visualizes the relationship between two categorical variables.
2. Easy to identify patterns or correlations in a tabular format.

### Cons:
1. May not be suitable for large datasets.
2. Requires preprocessing data into a matrix format.

```python
# Compute the correlation matrix
correlation_matrix = tips.corr(numeric_only=True)

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='rocket')
plt.title('Correlation Heatmap - Tips Dataset')
plt.show()
```

## Sankey Diagram
### Pros:
1. Illustrates flow and relationships between entities or stages.
2. Effective for displaying complex data with multiple paths.

### Cons:
1. Can become cluttered and hard to follow with excessive connections.
2. Requires careful design to prevent ambiguity in large datasets.

```python
fig=px.parallel_categories(tips, dimensions=['day', "sex", "smoker"],
                          color="size", color_continuous_scale=px.colors.sequential.Inferno,)
fig.show()
```

## Radar Chart
### Pros:
1. Compares multiple variables across different categories.
2. Highlights relative strengths and weaknesses of each category.

### Cons:
1. Can be complex to interpret with a large number of variables.
2. Requires careful scaling to avoid distorting the chart.

```python
fig = go.Figure()

categories = ["accuracy", "precision", "recall", "F1 score", "specificity"]

fig.add_trace(go.Scatterpolar(
        r=[4, 3, 2, 4, 2],
        theta=categories,
        fill='toself',
        name='Model 1'
    ))

fig.add_trace(go.Scatterpolar(
        r=[3, 2, 3, 2, 4],
        theta=categories,
        fill='toself',
        name='Model 2'
    ))

# Define layout for the radar chart
fig.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[0, 5]
    )),
  showlegend=False
)

# Display the chart
fig.show()
```

## Geographical Chart
### Pros:
1. Displays geographic data and patterns across regions.
2. Provides a visual representation of spatial distributions.

### Cons:
1. Requires geospatial data for accurate representations.
2. Can often lead to count plots if you aren't careful

```python
fig = px.choropleth(map_data, locations='iso_alpha', color='gdpPercap', hover_name='country', projection='natural earth')
fig.update_layout(title_text='Map')
fig.show()
```

## 3D Graph
### Pros:
1. Visualizes relationships in three-dimensional space.
2. Suitable for displaying complex data with multiple dimensions.

### Cons:
1. Can be challenging to interpret with overlapping or complex shapes.
2. Limited to displaying three variables at a time.

```python
fig = px.scatter_3d(iris, x='sepal_length', y='sepal_width', z='petal_length', color='species')
fig.update_layout(title_text='3D Graph')
fig.show()
```

# 1 : Labelling and Legends

```python
# Load data sets again
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

### Customising labels and legends

```python
color_map = {'setosa':'tab:blue', 'versicolor':'tab:red', 'virginica':'tab:green'}
for species in iris['species'].unique():
    species_data = iris[iris['species'] == species]
    plt.scatter(species_data['sepal_length'], species_data['sepal_width'], c=color_map[species], label=species)

plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize=12)
plt.title('Customizing Labels and Legends - Matplotlib', fontsize=14)
plt.legend(loc='upper right') # change legend location
plt.show()
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize=12)
plt.title('Customizing Labels and Legends - Seaborn', fontsize=14)
plt.legend(loc='upper right')
plt.show()
```

## Chart Styling

We can also style our plots by looking through the documentation we can see what features we can play with!<br>
In the kwargs section you can find additional parameters you can pass to the function such as:
- For Patches - https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Patch.html#matplotlib.patches.Patch
- For text - https://matplotlib.org/3.1.1/api/text_api.html#matplotlib.text.Text

A full list of matplotlib modules can be found here (https://matplotlib.org/3.1.1/api/index.html#modules) <br>
We will show an example with plt but keep in mind you can change a variety of features not limited to:

- Plot titles - plt.title
- Plot annotations - plt.annotate
- Axis labels - plt.xlabel, plt.ylabel
- Axis ranges - plt.xlim, plt.ylim
- Plot colors - facecolor, edgecolor

```python
counts, bins, patches = plt.hist(tips['total_bill'], bins=20, edgecolor='#1B1B1B', linewidth=0.4, alpha=0.87)
counts = counts.astype('int')
for i in range(len(patches)):
    patches[i].set_facecolor(plt.cm.viridis(counts[i]/max(counts)))

plt.title('Total Bill Distribution', fontsize=14, fontfamily='fantasy')
plt.xlabel('Bill ($)', fontsize=10, fontfamily='sans-serif')
plt.ylabel('Occurences (Frequency)', fontsize=10, fontfamily='sans-serif')
plt.show()
```

# 2 : Subplots

```python
# Load a dataset from seaborn package
tips = sns.load_dataset('tips')
```

```python
# Create subplots with a 2x2 grid (rows, columns)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot on the first subplot (top-left) - Scatter plot
# axs[row,col]
axs[0, 0].scatter(tips['total_bill'], tips['tip'], color='blue')
axs[0, 0].set_xlabel('Total Bill')
axs[0, 0].set_ylabel('Tip')
axs[0, 0].set_title('Scatter Plot')

# Plot on the second subplot (top-right) - Box plot
axs[0, 1].boxplot(tips['total_bill'], vert=False)
axs[0, 1].set_xlabel('Total Bill')
axs[0, 1].set_title('Box Plot')

# Plot on the third subplot (bottom-left) - Violin plot
axs[1, 0].violinplot(tips['tip'], vert=False)
axs[1, 0].set_xlabel('Tip')
axs[1, 0].set_title('Violin Plot')

# Plot on the fourth subplot (bottom-right) - Bar plot
axs[1, 1].bar(tips['day'], tips['total_bill'])
axs[1, 1].set_xlabel('Day')
axs[1, 1].set_ylabel('Total Bill')
axs[1, 1].set_title('Bar Plot')

# Adjust spacing between subplots
plt.tight_layout()

# Show the subplots
plt.show()
```
